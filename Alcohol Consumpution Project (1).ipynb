```python
#Here we perform different types of visualizaion like Scatterplot, lineplot, line plot  & Scatterplot,
#Histogram, countplot, pie graph and boxplot.
#for Visualization we first use inbuilt function Matplotlib and Seaborn
```

```python
import numpy as np #importing numpy pt is an alias of np
import pandas as pd #importing pandas is an alias ofpd
```

```python
Data=pd.read_csv("Drinks.csv")
```

```python
A = Data.head(10)
A
```

```python
A['country']
```

```python
import matplotlib.pyplot as plt #importing matplotlib plt is an alias of pyplot
import seaborn as sns #importing seaborn pt is an alias of sns
```

# Count Plot

```python
sns.countplot(data=A,x='beer_servings')
plt.title('Beer Serving in 2021')
plt.xticks(rotation=90)
plt.show
```

```python
sns.countplot(data=A,x='spirit_servings')
plt.title('Spirit Serving in 2021')
plt.xticks(rotation=90)
plt.show
```

```python
sns.countplot(data=A,x='wine_servings')
plt.title('Wine Serving in 2021')
plt.xticks(rotation=90)
plt.show
```

```python
sns.countplot(data=A,x='total_litres_of_pure_alcohol') 
plt.title('total_litres_of_pure_alcohol in 2021')
plt.xticks(rotation=90)
plt.show
```

# Scatter Plot

```python
#Creating scatter plot graph 
X=A["country"]
Y=A["beer_servings"]
plt.scatter(X,Y,color='red',marker='*',s=40)
plt.xlabel("country",color="red") # xlabel() defines the label of x-axis 
plt.ylabel("beer_servings",color="green") #ylabel() defines the label of x-axis
plt.title("Beer_servings in countries",color="Blue") #title() is used to give title of this scatter plot
plt.xticks(rotation=90)
plt.show()
```

```python
#Creating scatter plot graph 
X=A["country"]
Y=A["spirit_servings"]
plt.scatter(X,Y,color='orange',marker='*',s=40)
plt.xlabel("country",color="red") # xlabel() defines the label of x-axis 
plt.ylabel("spirit_servings",color="green") #ylabel() defines the label of x-axis
plt.title("spirit_servings in countries",color="Blue") #title() is used to give title of this scatter plot
plt.xticks(rotation=90)
plt.show()
```

```python
#Creating scatter plot graph 
X=A["country"]
Y=A["wine_servings"]
plt.scatter(X,Y,color='green',marker='*',s=40)
plt.xlabel("country",color="red") # xlabel() defines the label of x-axis 
plt.ylabel("wine_servings",color="green") #ylabel() defines the label of x-axis
plt.title(" wine_servings in countries",color="Blue") #title() is used to give title of this scatter plot
plt.xticks(rotation=90)
plt.show()
```

# Line plot and Scatter plot 

```python
#here we plot the graph by using variable country and beer servings
X=A["country"] 
Y=A["beer_servings"]
plt.plot(X,Y,color="yellow")
plt.scatter(X,Y,color="red",marker="*",s=60) #here we plot graph by using * marker and its s is the size 
plt.xlabel("countries",color="green")# xlabel() defines the label of x-axis 
plt.ylabel("beer_servings",color="green")# ylabel() defines the label of y-axis 
plt.xticks(rotation=90)
plt.show()
```

```python
#here we plot the graph by using variable country and wine servings
X=A["country"] 
Y=A["wine_servings"]
plt.plot(X,Y,color="green")
plt.scatter(X,Y,color="red",marker="*",s=60) #here we plot graph by using * marker and its s is the size 
plt.xlabel("countries",color="green")# xlabel() defines the label of x-axis 
plt.ylabel("wine_servings in India",color="green")# ylabel() defines the label of y-axis 
plt.xticks(rotation=90)
plt.show()
```

```python
#here we plot the graph by using variable country and spirit servings
X=A["country"] 
Y=A["spirit_servings"]
plt.plot(X,Y,color="orange")
plt.scatter(X,Y,color="red",marker="*",s=60) #here we plot graph by using * marker and its s is the size 
plt.xlabel("countries",color="green")# xlabel() defines the label of x-axis 
plt.ylabel("spirit_servings",color="green")# ylabel() defines the label of y-axis 
plt.xticks(rotation=90)
plt.show()
```

# Pie Graph

```python
#Pie graph in country
A["country"].value_counts().plot(kind="pie",autopct="%.01f%%")
plt.show()
```

```python
#pie graph of total liters alcohol
A["total_litres_of_pure_alcohol"].value_counts().plot(kind="pie",autopct="%.01f%%")
plt.show()
```

# Bar Plot

```python
#Here we show the representation of bar plot  between Country Wise and total liters of pure alcohol
plt.bar(A["country"],A["total_litres_of_pure_alcohol"],color=["red","green"])
plt.xlabel("Countries",color="green")# xlabel() defines the label of x-axis 
plt.ylabel("total_litres_of_pure_alcohol",color="maroon")# ylabel() defines the label of y-axis 
plt.title("Total_litres_of_pure_alcohol in Countries",color="blue") #title() is used to give title of this bar plot
plt.xticks(rotation=90)
plt.show()
```
